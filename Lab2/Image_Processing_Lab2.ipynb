# Lab Sheet-2:Image Enhancement Techniques

* OBJECTIVES: Apply Contrast and brightness enhancement
* TASK: Use histogram Equalization and contrast stretching
* ACTIVITY: compare results on different image datasets

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
image_path = "/content/conteastimage.jpg"
img = cv2.imread(image_path, 0)

# Histogram Equalization
hist_eq = cv2.equalizeHist(img)

# Define Contrast Stretching Function
def contrast_stretching(image):
    min_val = image.min()
    max_val = image.max()
    stretched = (image - min_val) * (255 / (max_val - min_val))
    return stretched.astype('uint8')

# Apply Contrast Stretching
contrast_stretched = contrast_stretching(img)

# Plotting results
titles = ['Original', 'Histogram Equalized', 'Contrast Stretched']
images = [img, hist_eq, contrast_stretched]

for i in range(3):
    plt.subplot(1, 3, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```

```python
#Histogram for comparision
plt.figure(figsize=(12,6))
for i in range(3):
  plt.subplot(1,3,i+1)
  plt.hist(images[i].ravel(),256,[0,256])
  plt.title(titles[i]+ "Histogram")
  plt.show()
```
